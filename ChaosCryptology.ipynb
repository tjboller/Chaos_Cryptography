```python
s = 30
r = 50
b = 5
v0 = [1.0,2.0,3.0] 
tf = 5000.0
t = np.linspace(0,tf,len(song))
```

```python
def lorenz(v,t):
    (x,y,z) = v
    x_dot = s * (y-x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return (x_dot,y_dot,z_dot)
```

```python
pos = odeint(lorenz,v0,t)
x = pos[:,0] + song
y = pos[:,1]
z = pos[:,2]
```

```python
write("Crypt.wav",sample_rate,x)
!afplay Crypt.wav
```

```python
fig = figure(figsize=(16,10))
ax = fig.gca(projection='3d')
ax.plot(x, y, z)
```

```python
def lorenz_sync(v,x,t):
    (xs,ys,zs) = v
    xs_dot = s * (ys-xs)
    ys_dot = r*x - ys - x*zs
    zs_dot = x*ys - b*zs
    return (xs_dot,ys_dot,zs_dot)
```

```python
dt = tf/len(song)

# Need one more for the initial values
xs = np.empty((len(t)))
ys = np.empty((len(t)))
zs = np.empty((len(t)))

# Setting initial values
xs[0], ys[0], zs[0] = (1., 2., 3.)


# Stepping through "time".
for i in range(len(x)-1) :
    # Derivatives of the X, Y, Z state
    xs_dot, ys_dot, zs_dot = lorenz((xs[i], ys[i], zs[i]),t)
    xs[i + 1] = xs[i] + (xs_dot * dt)
    ys[i + 1] = ys[i] + (ys_dot * dt)
    zs[i + 1] = zs[i] + (zs_dot * dt)
    
```

```python
def lorenz_song(xr,yr,zr,x,t):
    xr_dot = s * (yr-xr)
    yr_dot = r*x - yr - x*zr
    zr_dot = x*yr - b*zr
    return (xr_dot,yr_dot,zr_dot)
```

```python
dt = tf/len(song)

# Need one more for the initial values
xr = np.empty((len(t)))
yr = np.empty((len(t)))
zr = np.empty((len(t)))

# Setting initial values
xr[0], yr[0], zr[0] = (1., 3., 3.05)


# Stepping through "time".
for i in range(len(x)-1) :
    # Derivatives of the X, Y, Z state
    xr_dot, yr_dot, zr_dot = lorenz_song(xr[i], yr[i], zr[i], message[i],t)
    xr[i + 1] = xr[i] + (xr_dot * dt)
    yr[i + 1] = yr[i] + (yr_dot * dt)
    zr[i + 1] = zr[i] + (zr_dot * dt)
    
```

```python
fig, ax = plt.subplots();
fig.set_size_inches(15,10);
t = np.linspace(0,len(song)/sample_rate,len(song))
ax.plot(t,message-xr)
```
